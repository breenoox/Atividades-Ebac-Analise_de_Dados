## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.relplot(data=data, x='carat', y='price', hue='clarity', style='clarity')
diamonds.set(title='Diamantes: Quilates vs Preço - Baseado na Clareza', xlabel='Quilates', ylabel='Preço (USD)')
diamonds._legend.set_title("Clareza")
```

**Insight do gráfico 1**: O preço de diamantes com menor peso em quilates tende a ser mais elevado quando sua clareza é superior, diamantes com menor clareza geralmente possuem um peso maior em quilates para igualarem seus valores.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2


flights = data.query('month == "Dec"').reset_index()

grafico = sns.barplot(data=flights, x='year', y='passengers', palette='dark')
grafico.set(title='Passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros')
```

**Insight do gráfico 2**: De acordo com as ánalises e com que o gráfico transmite, observa-se uma crescente de passageiros no mês de dezembro ao longo dos anos,

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query('1949 <= year < 1960').reset_index()

grafico = sns.barplot(data=flights, x='year', y='passengers', hue='month', palette='bright')
grafico.set(title='Passageiros por mês entre 1949 e 1959', xlabel='Ano', ylabel='Passageiros')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

**Insight do gráfico 3**: Conseguimos observar que ao longo dos anos, todos os meses tiveram uma crescente de passsageiros, porém, os meses Julho e Agosto se destacam do começo ao fim.
